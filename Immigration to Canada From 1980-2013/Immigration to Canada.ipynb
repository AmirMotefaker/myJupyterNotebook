Sorting and Transpose

```python
# Sorting Data and creating new data set for top 5
years = list(map(str,range(1980, 2014)))
df_can.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()
```

Area Graph

```python
# Genrating Area Plot Of Top 5 immigrant Countries from 1980 to 2013
df_top5.plot(kind = "area", figsize = (15,8))

plt.title("Immigration Trend Of Top 5 Countries")
plt.ylabel("Number Of Immigrants")
plt.xlabel("Years")

plt.show()
```

Histogram

```python
# Plotting Histogram Without Alignment
df_can['2013'].plot(kind = 'hist', figsize =(10, 5))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# Plotting histogram with numpy alignment
import numpy as np

count, bin_edges = np.histogram(df_can['2013']) 
df_can['2013'].plot(kind = 'hist', xticks = bin_edges, figsize = (10, 5))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

Bar Graph

```python
df_iceland = df_can.loc['Iceland', years]
```

```python
df_iceland.plot(kind = 'bar', figsize = (10, 5))

plt.title('Icelandic Immigration to Canada from 1980 to 2013')
plt.ylabel('Numbers Of Immigrants')
plt.xlabel('Years')

plt.show()
```
